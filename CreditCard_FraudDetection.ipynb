```python
credit_cols = ['V11', 'V4', 'V2', 'V14', 'V17', 'V3', 'V12', 'V10', 'V6', 'V16']
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(18, 7))

# Iterate using enumerate to get both the index and the column name
for index, col in enumerate(credit_cols):
    # Calculate the row (r) and column (c) index for the axes array
    # Since there are 5 columns, index // 5 gives the row (0 or 1)
    # index % 5 gives the column (0 to 4)
    r = index // 5
    c = index % 5
    axes[r, c].scatter(x=credit[col], y=credit['Class'])

    # Set title and labels directly on the subplot object
    axes[r, c].set_title(f'{col} vs Class', fontsize=10)
    axes[r, c].set_xlabel(col, fontsize=9)
    axes[r, c].set_ylabel('Class', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
#Try to fiind relation between Amount and Fraud
plt.scatter(x=credit['Amount'], y=credit['Class'])
# Not very useful column top predict the Result
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(credit.drop('Class', axis=1), credit['Class'], test_size=0.2, random_state=42)
#
```

#WITHOUNT FEATURE SELECTION OR EXTRACTION
(Logistic Regression and Desicion Tree)

```python
# By Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model = lr.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred , y_test)
```

```python
#By Desicion Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model = dt.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_pred , y_test)
```

Here Logistic Regression is slightly better than Desicion tree but not very much.

# NOW I AM TRYING TO CHECH THAT WHAT COLUMN IS MORE IMPORTANT FOR THE CLASSIFICATION.

```python
lr_score =[]

for i in range (1,29):
  df= credit[[f'V{i}',  'Class']]
  dfx_train,dfx_test,dfy_train,dfy_test=train_test_split(df.drop('Class',axis=1),df['Class'],test_size=0.2,random_state=42)
  lr_model_df = lr.fit(dfx_train,dfy_train)
  dfy_pred=lr_model_df.predict(dfx_test)
  score= accuracy_score(dfy_pred,dfy_test)
  lr_score.append((f'V{i}', score))

lr_score_df = pd.DataFrame(lr_score, columns=['Feature', 'Accuracy'])
lr_score_df = lr_score_df.sort_values(by='Accuracy', ascending=False) # Sort the DataFrame
print(lr_score_df)
```

#**CONCLUSION OF FEATURE SELECTION:**   I get the top columns has maximum accuracy which I selected previously by seeing the corealtion between columns and class.

```python
pca.explained_variance_ratio_
```
