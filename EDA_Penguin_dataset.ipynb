```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasets = print(sns.get_dataset_names())
# Now For Day 01 the dataset is peguins
df1 = sns.load_dataset('penguins')
df1.head()
```

```python
# Identify rows with any missing values
rows_with_missing_values = df1[df1.isnull().any(axis=1)]
display(rows_with_missing_values)
```

#Univarient Analysis

```python
g = sns.displot(data=df1, x='bill_length_mm', kde=True, col='island', kind='hist')
g.set_axis_labels('Bill Length (mm)', 'Frequency')
g.set_titles(col_template='{col_name} Island')
g.fig.suptitle('Distribution of Bill Length (mm) by Island', y=1.02)
plt.tight_layout()
plt.show()
```

```python
g = sns.displot(data=df1, x='bill_depth_mm', kde=True, col='island', kind='hist')
g.set_axis_labels('Bill Depth (mm)', 'Frequency')
g.set_titles(col_template='{col_name} Island')
g.fig.suptitle('Distribution of Bill Depth (mm) by Island', y=1.02)
plt.tight_layout()
plt.show()
```

```python
g = sns.displot(data=df1, x='flipper_length_mm', kde=True, col='island', kind='hist')
g.set_axis_labels('Flipper Length (mm)', 'Frequency')
g.set_titles(col_template='{col_name} Island')
g.fig.suptitle('Distribution of Flipper Length (mm) by Island', y=1.02)
plt.tight_layout()
plt.show()
```

```python
g = sns.displot(data=df1, x='body_mass_g', kde=True, col='island', kind='hist')
g.set_axis_labels('Body Mass (g)', 'Frequency')
g.set_titles(col_template='{col_name} Island')
g.fig.suptitle('Distribution of Body Mass (g) by Island', y=1.02)
plt.tight_layout()
plt.show()
```

```python
df1['species'].value_counts()
```

```python
plt.figure(figsize=(24, 8))

# Create a mask for the upper triangle of the correlation matrix
mask = np.triu(Adelie.corr(numeric_only=True))

# For Adelie
plt.subplot(1, 3, 1)
sns.heatmap(Adelie.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, mask=mask)
plt.title('Adelie Data Correlation', fontsize=14)

# For Gentoo
plt.subplot(1, 3, 2)
sns.heatmap(Gentoo.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, mask=mask)
plt.title('Gentoo Data Correlation', fontsize=14)

# For Chinstrap
plt.subplot(1, 3, 3)
sns.heatmap(Chinstrap.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, mask=mask)
plt.title('Chinstrap Data Correlation', fontsize=14)

plt.tight_layout()
plt.show()
```
